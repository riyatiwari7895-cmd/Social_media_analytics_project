# Instagram Analytics: Data Cleaning and Preparation

## Introduction
Before diving into complex analysis or SQL queries, it is crucial to ensure our data is clean, formatted correctly, and ready for exploration. In this notebook, we will load the raw `Instagram_Analytics.csv` dataset, inspect its data types, handle any missing values, and engineer a few foundational features (like extracting temporal data from the datetime column).

```python
import pandas as pd
import numpy as np

# Set display options to view all columns easily
pd.set_option('display.max_columns', None)
```

## 1. Load the Raw Data
We will load the dataset from our `data/raw/` directory.

```python
# Adjust the path if your structure is different
df = pd.read_csv(r'../data/raw/Instagram_Analytics.csv')
df.head()
```

## 2. Initial Inspection
Let's check the data types, dimensions, and look for missing values.

```python
print("Dataset Shape:", df.shape)
print("\nData Types and Missing Values:")
df.info()
```

```python
# Check for explicit null values
df.isnull().sum()
```

```python
# Check for duplicated rows based on post_id
duplicates = df.duplicated(subset=['post_id']).sum()
print(f"Duplicate post_ids found: {duplicates}")
```

## 3. Data Type Formatting & Feature Engineering
The `post_datetime` column should be a datetime object so we can extract trends over time. We will also calculate an `interaction_rate` to complement the existing `engagement_rate`.

```python
# Convert datetime column to pandas datetime objects
df['post_datetime'] = pd.to_datetime(df['post_datetime'])
df['post_date'] = pd.to_datetime(df['post_date'])

# Feature Engineering: Extract Month and Year
df['post_month'] = df['post_datetime'].dt.month
df['post_year'] = df['post_datetime'].dt.year

# Calculate total interactions (likes + comments + shares + saves)
df['total_interactions'] = df['likes'] + df['comments'] + df['shares'] + df['saves']

df.head()
```

## 4. Export Processed Data
Now that the data is clean and enriched, we export it to our `processed` folder for EDA and SQL integration.

```python
df.to_csv('../data/processed/cleaned_instagram_data.csv', index=False)
print("Cleaned dataset saved successfully!")
```
