# Instagram Analytics: Exploratory Data Analysis (EDA)

## Introduction
With our data cleaned, this notebook focuses on discovering trends, patterns, and insights. We will visualize how different variables (like media type, account type, and post timing) affect engagement rates, reach, and performance buckets.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set_theme(style="whitegrid")
```

```python
# Load the cleaned dataset
df = pd.read_csv('../data/processed/cleaned_instagram_data.csv')
df['post_datetime'] = pd.to_datetime(df['post_datetime']) # Re-parse datetime
df.head()
```

## 1. Distribution of Performance Buckets
Let's see how our target variable (`performance_bucket_label`) is distributed across the dataset.

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='performance_bucket_label', order=['low', 'medium', 'high', 'viral'], palette='viridis')
plt.title('Distribution of Post Performance')
plt.ylabel('Number of Posts')
plt.xlabel('Performance Bucket')
plt.show()
```

## 2. Engagement by Account Type and Media Type
Does being a creator versus a brand change how different media types (reels, carousels, images) perform?

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='media_type', y='engagement_rate', hue='account_type', palette='muted')
plt.title('Average Engagement Rate by Media and Account Type')
plt.ylabel('Engagement Rate')
plt.xlabel('Media Type')
plt.show()
```

## 3. Correlation Matrix
Let's look at the linear correlation between numerical metrics like likes, shares, caption length, and follower count.

```python
numerical_cols = ['follower_count', 'likes', 'comments', 'shares', 'saves', 'reach', 'caption_length', 'hashtags_count', 'engagement_rate']
corr_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
